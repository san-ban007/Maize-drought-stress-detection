```python
import matplotlib.pyplot as plt
import pandas as pd
import matplotlib as mpl
import numpy as np
import matplotlib.colors
```

```python
trial_num ='008'
plant = 'Plant3'
Type='YEL_Stem'
#day = 'y20m10d25'
src_dir='/home/sbaner24/Maize/Output/NIR_Histograms/Trial'

folder = str(src_dir)+str(trial_num)+'/'+str(Type)+'/'+str(plant)+'/'  # path to folder having one histogram per sessions saved as csv files

fileList = Dates = ['y20m12d14','y20m12d15','y20m12d16','y20m12d17','y20m12d18','y20m12d19','y20m12d20','y20m12d21','y20m12d22']
x = np.linspace(0, 10, 100)
n_lines=11
c = np.arange(1., n_lines + 1)

cmap = plt.get_cmap("autumn_r", len(c))                                         # color theme for plots
norm = matplotlib.colors.BoundaryNorm(np.arange(len(c)+1)+0.5,len(c))           
sm = plt.cm.ScalarMappable(norm=norm, cmap=cmap)
sm.set_array([])  

fig, ax = plt.subplots(dpi=100)
plt.ylim([0,0.08])
for i in range(0,len(fileList)):

    df = pd.read_csv(folder + fileList[i]+'.csv')                                                  
    ax.plot(range(len(df)),df,c=cmap(i),linewidth = 3)

plt.xticks(fontsize = 10)
plt.yticks(fontsize = 10)
plt.xlabel('Pixels', fontsize = 10)
plt.ylabel('Normalized pixel count', fontsize = 10)
x = fig.colorbar(sm, ticks=c)
x.ax.tick_params(labelsize=10)
plt.show()
```

```python
# Importing libraries
import matplotlib.pyplot as plt
import numpy as np
import math

trial_num ='008'
plant1 = 'Plant1'
plant2 = 'Plant2'
plant3 = 'Plant3'
plant4 = 'Plant4'
Type='YEL_Stem'
#day = 'y20m10d25'
src_dir='/home/sbaner24/Maize/Output/NIR_Histograms/Trial'

folder2 = str(src_dir)+str(trial_num)+'/'+str(Type)+'/'+str(plant2)+'/'
folder3 = str(src_dir)+str(trial_num)+'/'+str(Type)+'/'+str(plant3)+'/'

#fileList =  ['y20m11d03','y20m11d04','y20m11d05','y20m11d06','y20m11d10','y20m11d11','y20m11d12','y20m11d13','y20m11d14']
#fileList =  ['y20m09d27','y20m09d28','y20m09d29','y20m09d30','y20m10d01','y20m10d02','y20m10d03','y20m10d04','y20m10d05','y20m10d06','y20m10d07','y20m10d08','y20m10d09','y20m10d10','y20m10d11','y20m10d12','y20m10d13']
#Mean Plot


y_mean2=[]
for day in fileList:
    with open(folder2+str(day)+'.txt') as f:
        line=f.readlines()
        line_new = [sub.replace('na', '0') for sub in line]
        y_mean2.append(float(line_new[0][5:-2]))
print('y_mean2=',y_mean2)

y_mean3=[]
for day in fileList:
    with open(folder3+str(day)+'.txt') as f:
        line=f.readlines()
        line_new = [sub.replace('na', '0') for sub in line]
        y_mean3.append(float(line_new[0][5:-2]))
print('y_mean3=',y_mean3)


plt.figure(figsize=(8,5))
plt.plot(fileList,y_mean2,'--o', color='r', label='Plant2(Drought Stressed)')
plt.plot(fileList,y_mean3,'-o', color='b', label='Plant3(Well Watered)')
plt.xticks(rotation = 20)
plt.xlabel("Day")
plt.ylabel("Mean")    
plt.legend()
plt.show()
```
